```python
def run_program(perms, vals):
    
    finalout = []
    for n in range(len(perms)):
    #for n in range(len(perms)):
        print(f'starting {n} permutation out of {len(perms)}')
        output4 = [perms[n][0], 0]
        output0  = [perms[n][1]]
        output1 = [perms[n][2]]
        output2 = [perms[n][3]]
        output3 = [perms[n][4]]
        outputs = [output0, output1, output2, output3, output4]
        inputs =[outputs[4], outputs[0], outputs[1], outputs[2], outputs[3]]
        
        vals_all = []
        for n in range(5):
            vals_all.append(vals.copy())
            
        threads = []
        for n in range(5):
            x = threading.Thread(target=run_amp, args=(vals_all[n], inputs[n], outputs[n]))
            threads.append(x)
            x.start()
            
        threads[0].join()
        threads[1].join()
        threads[2].join()
        threads[3].join()
        threads[4].join()
        #for index, thread in enumerate(threads):
            
        #    thread.join()
        
        
        #ot = handle_op(vals, perms, inputs, outputs)
        finalout.append(output4[-1])
        
    return np.max(finalout)
```

```python
def handle_op(code, vals, inputs, outputs, cnt, idx):
    opcode = int(code[-2:])
    idx1 = idx[0]
    #print(f'initial: {idx}')
    #print(f'changed to idx1: {idx1}')
    if opcode == 1:
        modes = code.zfill(5)
        modes = modes[0:-2]
        #vals = incodem(vals, idx, opcode, modes[0:-2])
        m1 = int(modes[-1])
        m2 = int(modes[-2])
            #m3 = int(modes[-3])
        if m1 == 0:
            pos1 = vals[idx1+1]
            #pdb.set_trace()
            
            v1 = vals[pos1]
            #print(f'm1 v1: {v1}')
        else:
            
            v1 = vals[idx1+1]
            #print(f'm1 v1: {v1}')
        if m2 == 0:
            pos2 = vals[idx1+2]
            v2 = vals[pos2]
        else:
            v2 = vals[idx1+2]

        pos3 = vals[idx1+3]           
        inc = 4
        idx1 += inc
        idx[0] = idx1
        #print(f'v1: idx')
        #print(f'v1: {v1}')
        #print(f'v2: {v2}')
        sval = v1+v2
        vals[pos3] = sval

        
    elif opcode == 2:
        modes = code.zfill(5)
        modes = modes[0:-2]
        m1 = int(modes[-1])
        m2 = int(modes[-2])
            #m3 = int(modes[-3])
        if m1 == 0:
            pos1 = vals[idx1+1]
            #pdb.set_trace()
            v1 = vals[pos1]
        else:
            v1 = vals[idx1+1]
        if m2 == 0:
            pos2 = vals[idx1+2]
            v2 = vals[pos2]
        else:
            v2 = vals[idx1+2]

        pos3 = vals[idx1+3]

        inc = 4
        idx1 += inc
        idx[0] = idx1
        sval = v1*v2
        vals[pos3] = sval

        #return idx
    elif opcode == 3:
        #pdb.set_trace()
        #print('asking for input 3')
        #print(f'vals[idx1+1] {vals[idx1+1]}')
        pos = vals[idx1+1]
        cnt1 = cnt[0]
        while len(inputs) < cnt1+1:
            time.sleep(.01)
        #print(f'input is {inputs[cnt1]}')
        vals[pos] = inputs[cnt1]
        cnt1 += 1
        cnt[0] = cnt1
        inc = 2
        idx1 += inc
        idx[0] = idx1

        #return idx
    elif opcode == 4:
        modes = code.zfill(3)
        mode0 = int(modes[0])
        if mode0 == 0:
            pos = vals[idx1+1]
            valr = vals[pos]
            #valr = incode4(vals, idx)
        else:
            valr = vals[idx1+1]

        inc = 2
        idx1 += inc
        idx[0] = idx1
        outputs.append(valr)
        #print(f'output is {valr}')
    elif opcode == 5:
        modes = code.zfill(4)
        modes = modes[0:-2]
        m1 = int(modes[-1])
        m2 = int(modes[-2])

        if m1 == 0:
            pos1 = vals[idx1+1]
            #pdb.set_trace()
            v1 = vals[pos1]
        else:
            v1 = vals[idx1+1]
        if m2 == 0:
            pos2 = vals[idx1+2]
            v2 = vals[pos2]
        else:
            v2 = vals[idx+2]


        if v1 != 0:
            idx1 = v2
            idx[0] = v2
            #move_forward = False
        else:
            idx1 += 3
            idx[0] = idx1

        #return idx[0]
    elif opcode == 6:
        modes = code.zfill(4)
        modes = modes[0:-2]

        m1 = int(modes[-1])
        m2 = int(modes[-2])

        if m1 == 0:
            pos1 = vals[idx1+1]
            #pdb.set_trace()
            v1 = vals[pos1]
        else:
            v1 = vals[idx1+1]
        if m2 == 0:
            pos2 = vals[idx1+2]
            v2 = vals[pos2]
        else:
            v2 = vals[idx1+2]


        if v1 == 0:
            idx1 = v2
            idx[0] = v2
            #move_forward = False
        else:
            idx1 += 3
            idx[0] = idx1

        #return idx
    elif opcode == 7:
        modes = code.zfill(5)
        modes = modes[0:-2]
        
        m1 = int(modes[-1])
        m2 = int(modes[-2])
            #m3 = int(modes[-3])
        if m1 == 0:
            pos1 = vals[idx1+1]
            #pdb.set_trace()
            v1 = vals[pos1]
        else:
            v1 = vals[idx1+1]
        if m2 == 0:
            pos2 = vals[idx1+2]
            v2 = vals[pos2]
        else:
            v2 = vals[idx1+2]

        pos3 = vals[idx1+3]

        if v1 < v2:
            vals[pos3] = 1
        else:
            vals[pos3] = 0
        inc = 4
        idx1 += inc
        idx[0] = idx1

        #
    elif opcode == 8:
        modes = code.zfill(5)
        modes = modes[0:-2]
        
        m1 = int(modes[-1])
        m2 = int(modes[-2])
            #m3 = int(modes[-3])
        if m1 == 0:
            pos1 = vals[idx1+1]
            #pdb.set_trace()
            v1 = vals[pos1]
        else:
            v1 = vals[idx1+1]
        if m2 == 0:
            pos2 = vals[idx1+2]
            v2 = vals[pos2]
        else:
            v2 = vals[idx1+2]

        pos3 = vals[idx1+3]

        if v1 == v2:
            vals[pos3] = 1
        else:
            vals[pos3] = 0

        inc = 4
        idx1 += inc
        idx[0] = idx1
        #return idx

    elif opcode == 99:
        return True
    return False
```
