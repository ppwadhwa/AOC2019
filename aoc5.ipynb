```python
def incodem(vals, idx, opcode, modes):
    m1 = int(modes[-1])
    m2 = int(modes[-2])
        #m3 = int(modes[-3])
    if m1 == 0:
        pos1 = vals[idx+1]
        #pdb.set_trace()
        v1 = vals[pos1]
    else:
        v1 = vals[idx+1]
    if m2 == 0:
        pos2 = vals[idx+2]
        v2 = vals[pos2]
    else:
        v2 = vals[idx+2]
    
    pos3 = vals[idx+3]
    
    if opcode == 1:
                    
        sval = v1+v2
        vals[pos3] = sval
        
        return vals
    
    elif opcode == 2:
            
        sval = v1*v2
        vals[pos3] = sval
        
        return vals
    elif opcode == 7:
       
        if v1 < v2:
            vals[pos3] = 1
        else:
            vals[pos3] = 0
            
    elif opcode == 8:
        if v1 == v2:
            vals[pos3] = 1
        else:
            vals[pos3] = 0
    
    return vals
    
```
