```python
def get_mode_value(vals, m1, idx1, rb):
    if m1 == 0:
        
        if vals[idx1] < len(vals):
            
            pos1 = vals[idx1]
            v1 = vals[pos1]
        else:
            #pdb.set_trace()
            v1 = 0

        
        #print(f'm1 v1: {v1}')
    elif m1 == 1:
        
        if idx1 < len(vals):
            
            v1 = vals[idx1]
        else:
            #pdb.set_trace()
            v1 = 0
    else:
        
        if vals[idx1]+rb < len(vals):
            
            pos1 = vals[idx1]+rb
            if pos1 < len(vals):
                
            #    zs = pos1 - len(vals) + 2
            #    vals.extend([0] * zs)
            
                v1 = vals[pos1]
            else:
                #pdb.set_trace()
                v1 = 0
        else:
            
            pos1 = 0
            v1 = vals[pos1]
        
        
    return v1
```

```python
def handle_direction(current_dir, turn, coords):
    #0 turns left, 1 turns right
    cx = coords[0]
    cy = coords[1]
    if current_dir == 'up':
        if turn == 1:
            current_dir = 'right'
            newx = cx+1
            newy = cy
        else:
            current_dir = 'left'
            newx = cx-1
            newy = cy
    elif current_dir == 'down':
        if turn == 1:
            current_dir = 'left'
            newx = cx-1
            newy = cy
        else:
            current_dir = 'right'
            newx = cx+1
            newy = cy
    elif current_dir == 'left':
        if turn == 1:
            current_dir = 'up'
            newx = cx
            newy = cy+1
        else:
            current_dir = 'down'
            newx = cx
            newy = cy-1
    elif current_dir == 'right':
        if turn == 1:
            current_dir = 'down'
            newx = cx
            newy = cy-1
        else:
            current_dir = 'up'
            newx = cx
            newy = cy+1
            
    #compute new coords
    
    return (newx, newy), current_dir
```
